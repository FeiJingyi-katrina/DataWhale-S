```python
import requests
import time
import re
from bs4 import BeautifulSoup
```

```python
def openProxyUrl(url):
    kv = {'user-agent' : 'Mozilla/5.0'}
    try:
        r = requests.get(url, headers = kv, timeout = 30)
        r.raise_for_status()
        # print(r.status_code)
        r.encoding = r.apparent_encoding
        return r.text
    except:
        print("")
```

```python
def getProxyIP(response):
    proxyIPList = []
    soup = BeautifulSoup(response, 'html.parser')
    proxyIPs = soup.find(id = 'ip_list').find_all('tr')
    for proxyIP in proxyIPs:
        if len(proxyIP.select('td')) >= 8:
            ip = proxyIP.select('td')[1].text
            port = proxyIP.select('td')[2].text # 端口
            protcool = proxyIP.select('td')[5].text # 类型
            if protcool in ('HTTP', 'HTTPS'):
                proxyIPList.append(f'{protcool}://{ip}:{port}')
    return proxyIPList
```

```python
def openUrlUsingProxy(url, proxy):
    kv = {'user-agent' : 'Mozilla/5.0'}
    proxies = {}
    if proxy.startswith('HTTPS'):
        proxies['https'] = proxy
    else:
        proxies['http'] = proxy
    try:
        r = requests.get(url, headers = kv, proxies = proxies, timeout = 30)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return (r.text, r.status_code)
    except:
        print("error url  " + url)
        print("error proxy  " + proxy)
        return False
```
