```python
import requests, json, re, random,time
from bs4 import BeautifulSoup
from selenium import webdriver
from lxml import etree
from copy import deepcopy
from selenium.common.exceptions import TimeoutException, NoSuchElementException
```

```python
def loginDXY(url):
    driver = webdriver.Chrome()
    try:
        driver.get(url)
        driver.maximize_window()
        driver.implicitly_wait(10)
        driver.find_element_by_link_text("登录").click()
        driver.find_element_by_link_text("返回电脑登录").click()
        driver.find_element_by_name("username").send_keys("18762866650")
        driver.find_element_by_name("password").send_keys("123zxcvbnm")
        driver.find_element_by_name("keepOnlineType").click()
        driver.find_element_by_class_name("button").click()
        time.sleep(10)
        print("成功")
        html = driver.page_source
        print(len(html))
        print(type(html))
    except TimeoutException:
        print("Time out")
        print("登录失败！")
    except NoSuchElementException:
        print("No Element")
        print("登录失败！")
    if html:
        print("抓取成功！")
        driver.quit()
        return html
    else:
        print("抓取失败！")
        driver.quit()
        return None
```

```python
def getUserInfo(url):
    try:
        kv = {'user-agent' : 'Mozilla/5.0'}
        r = requests.get(url, headers = kv)
        r.raise_for_status()  
        info = []
        tree = etree.HTML(r.text)
        info.append(t.xpath('//div[@class="banner-inner__user-id pa"]/a/text()')[0])
        try:
            info.append(t.xpath('//div[@class="user-level-area"]/text()')[0])
        except:
            info.append(t.xpath('//span[@class="level-wrap__level-title mr15"]/text()')[0])
        info.extend(t.xpath('//div[@class="follows-fans clearfix"]/div/p/a/text()'))
        info.extend(t.xpath('//div[@class="main-nav-inner"]/ul/li/span[1]/text()'))
        print(info)
        return info
    except:
        print("访问出错")
        return ""  
```

```python
def extractData(html):    
    tree = etree.HTML(html)
    ls_content = []
    ls = tree.xpath('//td[@class="postbody"]')
    length = len(ls)
    count = 0  
    for i in range(length):
        count += 1
        try:
            ls_content.append(''.join(ls[i].xpath('.//text()')).strip())  
        except:
            print('抓取第{}评论出错'.format(count))
            continue
    ls_urls = tree.xpath('//div[@class="auth"]/a/@href')
    ls_user_info = []
    n = 0
    for url in ls_urls:
        n += 1
        print("现在开始抓取第{}位用户的主页：{}".format(n, url))
        info = getUserInfo(str(url))
        ls_user_info.append(info)
    total = list(zip(ls_user_info, ls_content))
    print(total[0])
    print("恭喜你！成功抓取信息{}条！".format(len(total)))
    return total
```

```python
def saveData(total, fpath):
    n = 0
    with open(fpath, 'a', encoding='utf-8') as f: 
        for i in total:
            n += 1
            try:
                print("现在开始写入第{}位用户的信息".format(n))
                p = deepcopy(i[0])
                p.append(i[1])
                print(p) 
                s = ','.join(p) + '\n'
                f.write(s) 
            except:
                print("警告！第{}条信息写入出错！".format(n))
                continue
```

```python
def main():
    url = 'http://www.dxy.cn/bbs/thread/626626#626626'
    fpath = r'D:\code\task3\dingxiang.csv'
    html = loginDXY(url)
    total = extractData(html)
    saveData(total, fpath)
    print("成功结束程序！")
```

```python
def count_spend_time(func):
    start_time = time.perf_counter()
    func()
    end_time = time.perf_counter()
    time_dif = (end_time - start_time)
    second = time_dif%60
    minute = (time_dif//60)%60
    hour = (time_dif//60)//60
    print('spend ' + str(hour) + 'hours,' + str(minute) + 'minutes,' + str(second) + 'seconds')

if __name__ == '__main__':
    count_spend_time(main)
```
