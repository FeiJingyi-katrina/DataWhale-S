```python
from lxml import etree
from urllib import request
import requests
```

```python
def crow(i):
    url = 'https://movie.douban.com/top250?start=' + str(25*i)
    kv = {'User-Agent' : 'Mozilla/5.0'}
    r = requests.get(url, headers = kv)
    print(r.status_code)
    r.encoding = r.apparent_encoding
    html = r.text
    #print(html)
    html = etree.HTML(html) #转换成xpath能处理的对象
    datas = html.xpath('//ol[@class="grid_view"]/li')
    a = 0
    for data in datas:
        data_title = data.xpath('div/div[2]/div[@class="hd"]/a/span[1]/text()')
        print("No: " + str(i*25+a+1))
        print(data_title)
        with open('douban.txt','a',encoding='utf-8') as f:
            f.write("No: " + str(i*25+a+1) +  '  ')
            print(str(i*25+a+1))
            f.write(data_title[0] + '\n')
            print()
        a+=1
for i in range(10):
    crow(i)
```
