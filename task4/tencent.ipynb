```python
import time
import requests
import csv
from  selenium import webdriver
from bs4 import BeautifulSoup
```

```python
url = "https://news.qq.com/"
r = requests.get(url)
r.raise_for_status()
print(r.status_code)
r.encoding = r.apparent_encoding
html = r.text
try:
    soup = BeautifulSoup(html, 'html.parser')
    news = []
    for new in soup.select(".detail"):
        if len(new.select('h3')) > 0:
            href = new.select('a')[0]['href']
            title = new.select('a').text()
            news.append(title, href)
    print(news)
except:
    print("")
```

```python
driver=webdriver.Chrome()
driver.get("https://news.qq.com")
for i in range(1,100):
    time.sleep(2)
    driver.execute_script("window.scrollTo(window.scrollX, %d);"%(i*200))
html = driver.page_source
soup = BeautifulSoup(html, "lxml")
jxtits = soup.find_all("div", {"class" : "jx-tit"})[0].find_next_sibling().find_all("li")
print("index", " , ", "title", " , ","url")
for i,jxtit in enumerate(jxtits):
    try:
        text=jxtit.find_all("img")[0]["alt"]
    except:
        text=jxtit.find_all("div",{"class":"lazyload-placeholder"})[0].text
    try:
        url=jxtit.find_all("a")[0]["href"]
    except:
        print(jxtit)
    print(i+1,",",text,",",url)
```
